```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Importing the data
in_dt = pd.read_csv('C:/Users/jatin/Desktop/EastWestAirlinesCluster.csv')
in_dt
```

```python
# Check data for missing values
in_dt.info()
```

```python
# Check data for any other inconsistencies
in_dt.describe()
```

```python
# Removing the column "ID#"
stzd_dt = in_dt.iloc[:,1:12]
in_dt1 = in_dt.iloc[:,1:12]
in_dt2 = in_dt.iloc[:,1:12]
in_dt3 = in_dt.iloc[:,1:12]
in_dt1
```

```python
# Standardize the data so that all the varibales are in scale

from sklearn import preprocessing

stzd_dt[['Balance','Qual_miles','Bonus_miles','Bonus_trans','Flight_miles_12mo','Flight_trans_12','Days_since_enroll']] = preprocessing.StandardScaler().fit_transform(in_dt1[['Balance','Qual_miles','Bonus_miles','Bonus_trans','Flight_miles_12mo','Flight_trans_12','Days_since_enroll']])
stzd_dt[['cc1_miles','cc2_miles','cc3_miles']] = preprocessing.MinMaxScaler().fit_transform(in_dt1[['cc1_miles','cc2_miles','cc3_miles']])
stzd_dt.head()
```

```python
# Hierarchical Clustering with Euclidean distance and Wardâ€™s method.

from scipy.cluster import hierarchy

hrc_dt1 = hierarchy.linkage(stzd_dt, 'ward', 'euclidean')
plt.figure(figsize=(15,10)) 
plt.title("Dendrogram with Euclidean distance and Wardâ€™s method")
plt.axhline(y=75, color='r', linestyle='-')
ddg1 = hierarchy.dendrogram(hrc_dt1)
```

### As we can see from the Dendogram, the data can be divided into 3 inferable clusters based on Hierarchical clustering with Euclidean distance and Wardâ€™s method.

### 

```python
# Comparing cluster centroids

clusters = hierarchy.cut_tree(hrc_dt1, n_clusters=3)

data1=pd.DataFrame(clusters)
in_dt1["Cluster_ID"]=data1

cluster1 = in_dt1.loc[in_dt1["Cluster_ID"]==0]
cluster2 = in_dt1.loc[in_dt1["Cluster_ID"]==1]
cluster3 = in_dt1.loc[in_dt1["Cluster_ID"]==2]
print('\n Cluster 1 \n',cluster1)
print('\n Cluster 2 \n',cluster2)
print('\n Cluster 3 \n',cluster3)
```

```python
in_dt1.groupby("Cluster_ID").mean()
```

```python
from plotly.offline import init_notebook_mode
init_notebook_mode(connected = True)
import plotly.express as px


fig = px.histogram(in_dt1, x="Cluster_ID", nbins=3, text_auto=True, histnorm='percent')
fig.update_layout(bargap=0.2)
fig.update_xaxes(dtick=1)
fig.show()
```

### Based on above observations we can give following categories to each cluster: 
### 1) Non-Frequent / First time Flyers (Cluster_ID 0): This cluster has non frequent or firt time flyers as we can see that the number of balance miles is the lowest of all groups. And Flight miles and flight transactions in last 12 months is also the lowest. It has maximum number of customers, around 67.9%.
### 2) Members Credit card user Flyers (Cluster_ID 1): This cluster amounts for 30.7% of customers and comprises of customers who are not that frequent as flight transactions in last 12 months is low, but are avid user of frequent flyer credit card as they have high Balance miles and highest Bonus miles, accrued from non flight transactions on credit card.
### 3) Frequent Flyers (Cluster_ID 2): This cluster has very low percentage, 1.4% but still has highest miles balance and also have highest Flight miles and flight transactions in last 12 months. Thus, they are frequent flyers who also use frequent flyer credit card and rewards credit card in order to accrue more miles to redeem.

### To check the stability of clusters, remove a random 5% of the data (by taking a random sample of 95% of therecords), and repeat the analysis

```python
stzd_dt.shape
```

```python
s95_dt = stzd_dt.sample(frac=.95)
s95_dt.shape
```

```python
hrc_dt2 = hierarchy.linkage(s95_dt, 'ward', 'euclidean')
plt.figure(figsize=(15,10)) 
plt.title("Dendrogram with Euclidean distance and Wardâ€™s method - 95% random sample")
plt.axhline(y=75, color='r', linestyle='-')
ddg2 = hierarchy.dendrogram(hrc_dt2)
```

```python
clusters = hierarchy.cut_tree(hrc_dt2, n_clusters=3)

data2=pd.DataFrame(clusters)
in_dt2["Cluster_ID"]=data2

in_dt2.head()
```

### We get the similar Dendrogram for 95% sample and thus, can conclude that our cluster model is stable

### 

### Cluster all passengers again using k-means clustering. 

```python
# K-means clustering
from sklearn.cluster import KMeans
fit = KMeans(n_clusters=3, max_iter = 10, random_state=0).fit(stzd_dt)

# Determine number of clusters
Cluster_Variability = []
for i in range(1,6): 
    kmeans =  KMeans(n_clusters=i).fit(stzd_dt)
    Cluster_Variability.append(kmeans.inertia_)
#Plot the elbow graph
plt.figure(figsize=(20,10))
plt.plot(range(1,6),Cluster_Variability,marker='o')
plt.title('The Elbow Method Graph')
plt.xticks(range(1,6))
plt.xlabel('Number of clusters')
plt.ylabel('Within groups sum of squares')
plt.show()
```

### On observing the elbow graph above we can see that for clusters greater than 3 there is not much significant change in within groups sum of squares (Cluster Variability) and thus, we can conclude that dividing the dataset into 3 clusters is most feasible solution.
